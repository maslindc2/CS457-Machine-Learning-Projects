Importing Numpy, MatPlotLib, and Genetic Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
import GeneticAlg as ga
```

Equation to maximize

```python
def function(x):
    # function to find max of: f(x,y) = sin(pi * 10 * x + 10/(1+y^2)) + ln(x^2+y^2)
    return np.sin(np.pi * 10 * x[0] + 10/(1 + x[1]**2)) + np.log(x[0]**2 + x[1]**2)
```

Define hyperparameters

```python
bounds = [[3.0, 10.0], [4.0, 8.0]] # range for each x and y
numIterations = 50 # controls total number of iteration
numBits = 16 # controls number of bits per variable
numPopulation = 80 # controls population size
crossoverRate = 0.9 # controls high probability cross over value
mutationRate = .5/(float(numBits) * len(bounds)) # calculates mutation rate low probability value
```

Run the genetic algorithm with the above hyperparameters

```python
output = ga.geneticAlgorithm(function, bounds, numBits, numIterations, numPopulation, crossoverRate, mutationRate)
```

```python
resultsAllGenerations = list(np.concatenate(output[0]).flat)
resultsNewBest = list(np.concatenate(output[1]).flat)

allGenXY = ga.fetchPlotValues(resultsAllGenerations) # Index [0] is X values, Index [1] is y values
newBestXY = ga.fetchPlotValues(resultsNewBest) #Index [0] is X values Index [1] is Y values
```

```python
print(f"Number of Generations: {numIterations}")
print(f"Population size: {numPopulation}")
print(f"Mutation Probability: {mutationRate}")
print(f"Crossover Rate: {crossoverRate}")
```

Evolution of the f(x,y) value for all generations

```python
plt.plot(allGenXY[0], allGenXY[1], marker='.')
plt.title(f"Evolution of f(x,y) value for {str(numIterations)} generations.")
plt.xlabel('Generations (x)')
plt.ylabel('f(x,y)');
```

Evolution of new best f(x,y) values for all generations.

```python
plt.plot(newBestXY[0], newBestXY[1], marker='.')
plt.title(f"Evolution of best new f(x,y) value over the {str(numIterations)} generations. \n Max value found after {np.max(newBestXY[0])} generations.")
plt.xlabel('Generations (x)')
plt.ylabel('f(x,y)');
```

Results

```python
decoded = ga.decode(bounds, numBits, output[2])
print(f"Maximum value of f(x,y) is {output[3]}, with an x value of {decoded[0]}, and a y value of {decoded[1]}")
```
